1.- Supón que, para un juego, tienes que calcular el desglose mínimo en billetes y monedas de una 
cantidad exacta de euros. Hay billetes de 500, 200, 100, 50, 20, 10 y 5 € y monedas de 2 y 1 €. Por 
ejemplo, deseamos conocer el desglose de 434 €

```python
def breakdown_euros(amount):
    # List of tuples with (value, description) in descending order
    denominations = [
        (500, "500€ bills"),
        (200, "200€ bills"),
        (100, "100€ bills"),
        (50, "50€ bills"),
        (20, "20€ bills"),
        (10, "10€ bills"),
        (5, "5€ bills"),
        (2, "2€ coins"),
        (1, "1€ coins")
    ]
    
    print(f"Breakdown for {amount}€:")
    for value, description in denominations:
        if amount >= value:
            count = amount // value  # Number of bills/coins of this denomination
            amount %= value          # Remaining amount to be broken down
            print(f"{description}: {count}")

# Main execution
euro_amount = 434  
breakdown_euros(euro_amount)
```

la función random del módulo random). El juego consiste en que el usuario debe adivinar el número,

le dirá si el número es mayor o menor que el que se ha de adivinar. El programa debe terminar cuando 
el usuario acierta el número.

```python
import random

def guess_the_number():
    # Ask the user for the maximum value (N)
    max_value = int(input("Enter the maximum number (N): "))
    
    # Generate a random secret number between 1 and N
    secret_number = random.randint(1, max_value)
    
    guess = None
    while guess != secret_number:
        try:
            guess = int(input("Enter your guess: "))
        except ValueError:
            print("Invalid input. Please enter an integer.")
            continue

        if guess < secret_number:
            print("The secret number is greater than your guess.")
        elif guess > secret_number:
            print("The secret number is less than your guess.")
        else:
            print("Congratulations, you guessed the number!")
guess_the_number()
```

4.- Escribir un programa que lea un número positivo y escriba el triángulo de Floyd hasta el número 
leído. 

```python
def print_floyd_triangle(n):
    number = 1  # Start counting from 1
    row = 1     # Row counter

    # Continue printing until the current number exceeds n
    while number <= n:
        # Print 'row' numbers in the current line
        for _ in range(row):
            if number > n:
                break  # Stop if we've reached beyond n
            print(number, end=" ")
            number += 1
        print()  # Newline after each row
        row += 1  # Increase the row length for the next iteration

try:
    n = int(input("Enter a positive number: "))
    if n <= 0:
        print("Please enter a positive number.")
    else:
        print_floyd_triangle(n)
except ValueError:
    print("Invalid input. Please enter an integer.")
```

5.-Tienes una lista de cervezas con sus respectivas puntuaciones. El programa permite al usuario agregar puntuaciones para distintas cervezas, calcular el promedio de puntuaciones y mostrar la lista de cervezas ordenadas según su puntuación promedio.

```python
# Beer Rating Analyzer

beer_ratings = {
    "IPA": [8, 9, 7],
    "Stout": [9, 8],
    "Lager": [6, 7, 7]
}

def add_rating(beer, rating):
    if beer in beer_ratings:
        beer_ratings[beer].append(rating)
    else:
        beer_ratings[beer] = [rating]

def get_average(ratings):
    return sum(ratings) / len(ratings)

beer_name = input("Enter the name of the beer to rate: ").strip()
try:
    rating = float(input("Enter your rating (0-10): "))
    if rating < 0 or rating > 10:
        print("Rating must be between 0 and 10.")
    else:
        add_rating(beer_name, rating)
except ValueError:
    print("Invalid rating input.")

print("\nBeer ratings (sorted by average rating):")
averages = [(beer, get_average(ratings)) for beer, ratings in beer_ratings.items()]
averages.sort(key=lambda x: x[1], reverse=True)

for beer, avg in averages:
    print(f"{beer}: Average Rating = {avg:.2f}")
```

6.- El juego del ahorcado consiste en que un jugador (A) debe adivinar la palabra pensada por el 
jugador (B). Para ello debemos utilizar dos cadenas de caracteres: en la primera se guardará la palabra 
que ha pensado el jugador (B), y la segunda se actualizará en cada turno para contemplar los aciertos 
del jugador (A). Esta segunda cadena debe inicializarse a una cadena de caracteres de la misma 
longitud que la primera pero rellena con caracteres como: ‘_’ ó ‘*’. El jugador (A) tiene una cantidad 
máxima de fallos permitidos (elegido por el programador). Se considera fallo preguntar por un 
carácter que no se encuentra en la palabra pensada por el jugador (B). El jugador (A) gana la partida 
cuando acierta la palabra sin consumir el total de fallos permitidos. El jugador (B) gana cuando el 
jugador (A) consume todos los fallos permitidos

```python
def hangman():
    # Player B enters the secret word
    secret_word = input("Player B, please enter the secret word: ").strip()
    
    # Initialize the hidden word (using underscores) with the same length as the secret word
    hidden_word = ["_"] * len(secret_word)
    
    # Set the maximum number of allowed mistakes
    max_mistakes = 6
    mistakes = 0

    # A set to keep track of already guessed letters
    guessed_letters = set()

    # Clear the screen so Player A cannot see the secret word
    print("\n" * 50)
    
    print("Welcome to Hangman!")
    print("Secret word: " + " ".join(hidden_word))
    
    # Continue the game until the player runs out of allowed mistakes or the word is completely guessed
    while mistakes < max_mistakes and "".join(hidden_word) != secret_word:
        guess = input("Player A, please guess a letter: ").strip().lower()
        
        # Check that the guess is a single alphabetic character
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue

        # If the letter was already guessed, inform the player and continue
        if guess in guessed_letters:
            print("You already guessed that letter. Try a different one.")
            continue

        guessed_letters.add(guess)

        # Check if the guessed letter is in the secret word
        # We compare lowercase versions to allow for case insensitivity
        if guess in secret_word.lower():
            # Update the hidden word with the correctly guessed letter
            for index, letter in enumerate(secret_word.lower()):
                if letter == guess:
                    # Use the original character (preserving case) from secret_word
                    hidden_word[index] = secret_word[index]
            print("Good guess!")
        else:
            mistakes += 1
            print(f"Wrong guess! You have {max_mistakes - mistakes} attempts left.")
        
        print("Current word: " + " ".join(hidden_word))
    
    # Check the outcome of the game
    if "".join(hidden_word) == secret_word:
        print("Congratulations, Player A! You guessed the word!")
    else:
        print("Sorry, Player A, you've run out of attempts. Player B wins!")
        print("The secret word was:", secret_word)

hangman()
```

7.- Crear de manera automática un tablero de damas donde las posiciones ocupadas inicialmente en 
una partida contengan el valor 1 para las fichas de un jugador y 2 para el otro. Las posiciones 
desocupadas contendrán el valor 0.

```python
def create_checkers_board():
    board = []
    for row in range(8):
        board_row = []
        for col in range(8):
            # For the top three rows, place player 1's pieces on dark squares
            if row < 3 and (row + col) % 2 == 0:
                board_row.append(1)
            # For the bottom three rows, place player 2's pieces on dark squares
            elif row > 4 and (row + col) % 2 == 0:
                board_row.append(2)
            else:
                board_row.append(0)
        board.append(board_row)
    return board

def print_board(board):
    for row in board:
        print(" ".join(str(cell) for cell in row))

board = create_checkers_board()
print_board(board)
```

```python
import tkinter as tk
from tkinter import messagebox

# Global board settings
BOARD_SIZE = 8    # 8x8 chess board
CELL_SIZE = 60    # Each cell is 60x60 pixels

def mark_rook_moves(board, row, col):
    # Mark horizontal moves (all columns in the given row)
    for j in range(BOARD_SIZE):
        if j != col:
            board[row][j] = "X"
    # Mark vertical moves (all rows in the given column)
    for i in range(BOARD_SIZE):
        if i != row:
            board[i][col] = "X"

def mark_bishop_moves(board, row, col):
    # The four diagonal directions: (down-right, down-left, up-right, up-left)
    directions = [(1, 1), (1, -1), (-1, 1), (-1, -1)]
    for d_row, d_col in directions:
        i, j = row + d_row, col + d_col
        while 0 <= i < BOARD_SIZE and 0 <= j < BOARD_SIZE:
            board[i][j] = "X"
            i += d_row
            j += d_col

def draw_board(canvas, board):
    """Draws the chessboard on the given canvas using the board array.
       Squares with a marker (i.e. "X", "R", or "B") are labeled.
    """
    # Clear previous drawing
    canvas.delete("all")
    
    # Define chessboard colors
    light_color = "#F0D9B5"
    dark_color = "#B58863"
    
    for i in range(BOARD_SIZE):
        for j in range(BOARD_SIZE):
            # Determine the color of the square (alternating pattern)
            if (i + j) % 2 == 0:
                color = light_color
            else:
                color = dark_color
                
            # Calculate square coordinates
            x1 = j * CELL_SIZE
            y1 = i * CELL_SIZE
            x2 = x1 + CELL_SIZE
            y2 = y1 + CELL_SIZE
            
            # Draw the square
            canvas.create_rectangle(x1, y1, x2, y2, fill=color, outline="black")
            
            # If the board cell is not empty, draw the marker text in the center
            if board[i][j] != ".":
                canvas.create_text((x1+x2)/2, (y1+y2)/2,
                                   text=board[i][j],
                                   font=("Arial", 20),
                                   fill="red")

def show_moves():
    """Reads the input values, computes the board with possible moves,
       and updates the graphical chess board.
    """
    piece = piece_var.get().strip().lower()
    try:
        row = int(row_entry.get())
        col = int(col_entry.get())
    except ValueError:
        messagebox.showerror("Input Error", "Please enter integer values for row and column.")
        return

    if piece not in ("rook", "bishop"):
        messagebox.showerror("Input Error", "Invalid piece. Please choose either 'rook' or 'bishop'.")
        return

    if not (0 <= row < BOARD_SIZE and 0 <= col < BOARD_SIZE):
        messagebox.showerror("Input Error", "Row and column indices must be between 0 and 7.")
        return

    # Create an empty board (filled with ".")
    board = [["." for _ in range(BOARD_SIZE)] for _ in range(BOARD_SIZE)]
    
    # Mark moves according to the selected piece
    if piece == "rook":
        mark_rook_moves(board, row, col)
        board[row][col] = "R"  # Mark the rook's position
    elif piece == "bishop":
        mark_bishop_moves(board, row, col)
        board[row][col] = "B"  # Mark the bishop's position

    # Redraw the board on the canvas with updated moves
    draw_board(board_canvas, board)

# Create the main Tkinter window
root = tk.Tk()
root.title("Chess Piece Moves")

# Create a frame for inputs
input_frame = tk.Frame(root)
input_frame.pack(side=tk.TOP, pady=10)

# Input for piece type
tk.Label(input_frame, text="Piece (rook or bishop):").grid(row=0, column=0, padx=5, pady=5)
piece_var = tk.StringVar(value="rook")
piece_entry = tk.Entry(input_frame, textvariable=piece_var, width=10)
piece_entry.grid(row=0, column=1, padx=5, pady=5)

# Input for row index
tk.Label(input_frame, text="Row (0-7):").grid(row=1, column=0, padx=5, pady=5)
row_entry = tk.Entry(input_frame, width=10)
row_entry.grid(row=1, column=1, padx=5, pady=5)

# Input for column index
tk.Label(input_frame, text="Column (0-7):").grid(row=2, column=0, padx=5, pady=5)
col_entry = tk.Entry(input_frame, width=10)
col_entry.grid(row=2, column=1, padx=5, pady=5)

# Button to process the input and display the board
show_button = tk.Button(input_frame, text="Show Moves", command=show_moves)
show_button.grid(row=3, column=0, columnspan=2, pady=10)

# Create a Canvas widget to draw the chessboard
board_canvas = tk.Canvas(root, width=CELL_SIZE * BOARD_SIZE, height=CELL_SIZE * BOARD_SIZE)
board_canvas.pack(side=tk.BOTTOM, padx=10, pady=10)

# Initially, draw an empty chessboard
empty_board = [["." for _ in range(BOARD_SIZE)] for _ in range(BOARD_SIZE)]
draw_board(board_canvas, empty_board)

# Run the Tkinter event loop
root.mainloop()
```

9.- Juego de las 4 en línea: El juego de las cuatro en línea se juega en una estructura vertical de 7 
filas y 8 columnas en la que dos jugadores van colocando fichas en cada columna. Las fichas caen, 
por efecto de la gravedad, hasta que llegan al fondo de la estructura o hasta que descansan sobre otra 
ficha colocada antes. Se pide hacer un simulador de este juego en Python que permita jugar a dos 
jugadores y que compruebe tras cada jugada, si alguno ha hecho cuatro en raya. Para ello, es 
conveniente dividir el programa en diferentes subprogramas que se vayan encargando de cada una de 
las tareas necesarias

```python
# Connect Four (4 in a Row) Game
# Board: 7 rows x 8 columns

def create_board():
    """Creates and returns a 7x8 board filled with '.' to represent empty cells."""
    return [["." for _ in range(8)] for _ in range(7)]

def print_board(board):
    """Prints the board with row and column indices."""
    # Print column headers
    print("  " + " ".join(str(i) for i in range(8)))
    for idx, row in enumerate(board):
        print(str(idx) + " " + " ".join(row))
    print()  # Blank line for readability

def place_token(board, col, token):
    """
    Attempts to place a token in the specified column.
    Tokens "fall" to the lowest available row.
    Returns the row index where the token was placed or None if the column is full.
    """
    for row in range(len(board) - 1, -1, -1):
        if board[row][col] == ".":
            board[row][col] = token
            return row
    return None  # Column is full

def check_win(board, token):
    """
    Checks whether the given token has 4 in a row on the board.
    Looks for horizontal, vertical, and two diagonal (down-right and down-left) sequences.
    Returns True if a winning sequence is found, False otherwise.
    """
    rows = len(board)
    cols = len(board[0])
    
    # Check horizontal locations for win
    for r in range(rows):
        for c in range(cols - 3):
            if (board[r][c] == token and board[r][c + 1] == token and
                board[r][c + 2] == token and board[r][c + 3] == token):
                return True

    # Check vertical locations for win
    for r in range(rows - 3):
        for c in range(cols):
            if (board[r][c] == token and board[r + 1][c] == token and
                board[r + 2][c] == token and board[r + 3][c] == token):
                return True

    # Check diagonal down-right
    for r in range(rows - 3):
        for c in range(cols - 3):
            if (board[r][c] == token and board[r + 1][c + 1] == token and
                board[r + 2][c + 2] == token and board[r + 3][c + 3] == token):
                return True

    # Check diagonal down-left
    for r in range(rows - 3):
        for c in range(3, cols):
            if (board[r][c] == token and board[r + 1][c - 1] == token and
                board[r + 2][c - 2] == token and board[r + 3][c - 3] == token):
                return True

    return False

def board_full(board):
    """Returns True if there are no empty cells on the board (a draw), False otherwise."""
    for row in board:
        if "." in row:
            return False
    return True

# Main game loop for two players
def play_game():
    board = create_board()
    current_player = 1
    # Assign tokens to players: Player 1 uses 'X' and Player 2 uses 'O'
    tokens = {1: "X", 2: "O"}
    
    while True:
        print_board(board)
        try:
            col = int(input("Player {} ({}), choose a column (0-7): "
                            .format(current_player, tokens[current_player])))
        except ValueError:
            print("Invalid input. Please enter an integer between 0 and 7.\n")
            continue
        
        if col < 0 or col >= 8:
            print("Column must be between 0 and 7.\n")
            continue
        
        row = place_token(board, col, tokens[current_player])
        if row is None:
            print("Column is full, choose another column.\n")
            continue
        
        # Check for a win after the token is placed
        if check_win(board, tokens[current_player]):
            print_board(board)
            print("Player {} ({}) wins!".format(current_player, tokens[current_player]))
            break
        
        # Check for a draw if the board is full
        if board_full(board):
            print_board(board)
            print("It's a draw!")
            break
        
        # Switch players
        current_player = 2 if current_player == 1 else 1

play_game()
```

```python
import tkinter as tk
from tkinter import messagebox

# Board dimensions and cell size in pixels
ROWS = 7
COLS = 8
CELL_SIZE = 80  # Each cell is 80x80 pixels

class ConnectFourGUI:
    def __init__(self, master):
        self.master = master
        master.title("Connect Four")
        
        # Create the Canvas to draw the board (blue background is typical for Connect Four)
        self.canvas = tk.Canvas(master, width=COLS * CELL_SIZE, height=ROWS * CELL_SIZE, bg="blue")
        self.canvas.pack(padx=10, pady=10)
        self.canvas.bind("<Button-1>", self.handle_click)
        
        # Status label to indicate the current player's turn or the game result
        self.status_label = tk.Label(master, text="Player 1's Turn (Red)", font=("Helvetica", 14))
        self.status_label.pack(pady=5)
        
        # Reset button to restart the game
        self.reset_button = tk.Button(master, text="Reset Game", font=("Helvetica", 12), command=self.reset_game)
        self.reset_button.pack(pady=5)
        
        # Initialize game state variables
        self.board = self.create_board()
        self.current_player = 1  # Player 1 uses token 1 (red), Player 2 uses token 2 (yellow)
        self.game_over = False
        self.draw_board()
    
    def create_board(self):
        """Creates and returns a 7x8 board filled with 0 to represent empty cells."""
        return [[0 for _ in range(COLS)] for _ in range(ROWS)]
    
    def draw_board(self):
        """Draws the board on the canvas. Each cell is drawn as a circle."""
        self.canvas.delete("all")
        for row in range(ROWS):
            for col in range(COLS):
                # Margins so that the tokens do not touch the borders of the cell
                x1 = col * CELL_SIZE + 5
                y1 = row * CELL_SIZE + 5
                x2 = (col + 1) * CELL_SIZE - 5
                y2 = (row + 1) * CELL_SIZE - 5
                token = self.board[row][col]
                if token == 0:
                    color = "white"  # Empty cell
                elif token == 1:
                    color = "green"
                elif token == 2:
                    color = "yellow"
                self.canvas.create_oval(x1, y1, x2, y2, fill=color, outline="black")
    
    def handle_click(self, event):
        """Handles mouse clicks on the canvas. Determines the clicked column and places the token."""
        if self.game_over:
            return
        
        # Determine which column was clicked based on the x-coordinate
        col = event.x // CELL_SIZE
        if col < 0 or col >= COLS:
            return
        
        # Attempt to place the token in the chosen column
        row = self.place_token(col, self.current_player)
        if row is None:
            messagebox.showinfo("Invalid Move", "That column is full. Choose another column.")
            return
        
        # Update the graphical board
        self.draw_board()
        
        # Check if the current player has won
        if self.check_win(self.current_player):
            self.game_over = True
            self.status_label.config(text=f"Player {self.current_player} wins!")
            messagebox.showinfo("Game Over", f"Player {self.current_player} wins!")
            return
        
        # Check if the board is full (draw)
        if self.board_full():
            self.game_over = True
            self.status_label.config(text="It's a draw!")
            messagebox.showinfo("Game Over", "It's a draw!")
            return
        
        # Switch to the other player
        self.current_player = 2 if self.current_player == 1 else 1
        if self.current_player == 1:
            self.status_label.config(text="Player 1's Turn (Red)")
        else:
            self.status_label.config(text="Player 2's Turn (Yellow)")
    
    def place_token(self, col, player):
        """
        Places the player's token in the specified column, simulating gravity.
        Returns the row index where the token is placed, or None if the column is full.
        """
        for row in range(ROWS - 1, -1, -1):
            if self.board[row][col] == 0:
                self.board[row][col] = player
                return row
        return None
    
    def board_full(self):
        """Returns True if the board is completely filled (draw), False otherwise."""
        for row in self.board:
            if 0 in row:
                return False
        return True
    
    def check_win(self, player):
        """
        Checks if the given player has four tokens in a row.
        It checks horizontally, vertically, and in both diagonal directions.
        """
        # Check horizontal wins
        for row in range(ROWS):
            for col in range(COLS - 3):
                if (self.board[row][col] == player and
                    self.board[row][col+1] == player and
                    self.board[row][col+2] == player and
                    self.board[row][col+3] == player):
                    return True
        
        # Check vertical wins
        for row in range(ROWS - 3):
            for col in range(COLS):
                if (self.board[row][col] == player and
                    self.board[row+1][col] == player and
                    self.board[row+2][col] == player and
                    self.board[row+3][col] == player):
                    return True
        
        # Check diagonal down-right wins
        for row in range(ROWS - 3):
            for col in range(COLS - 3):
                if (self.board[row][col] == player and
                    self.board[row+1][col+1] == player and
                    self.board[row+2][col+2] == player and
                    self.board[row+3][col+3] == player):
                    return True
        
        # Check diagonal down-left wins
        for row in range(ROWS - 3):
            for col in range(3, COLS):
                if (self.board[row][col] == player and
                    self.board[row+1][col-1] == player and
                    self.board[row+2][col-2] == player and
                    self.board[row+3][col-3] == player):
                    return True
        
        return False
    
    def reset_game(self):
        """Resets the game state, clearing the board and starting a new match."""
        self.board = self.create_board()
        self.current_player = 1
        self.game_over = False
        self.draw_board()
        self.status_label.config(text="Player 1's Turn (Red)")

# Main execution: create the window and run the application
if __name__ == "__main__":
    root = tk.Tk()
    game = ConnectFourGUI(root)
    root.mainloop()
```

10.- Se requiere un sistema de gestión de inventario de cervezas que permita al usuario agregar cervezas, eliminar una cerveza, actualizar la cantidad disponible y listar el inventario completo. Se utiliza programación orientada a objetos para modelar cada cerveza y el inventario.

```python
# Beer Inventory Management System 

class Beer:
    def __init__(self, name, style, quantity, price):
        self.name = name
        self.style = style
        self.quantity = quantity
        self.price = price

    def __str__(self):
        return f"Name: {self.name}, Style: {self.style}, Quantity: {self.quantity}, Price: ${self.price:.2f}"

class Inventory:
    def __init__(self):
        self.beers = []

    def add_beer(self, beer):
        self.beers.append(beer)
        print(f"{beer.name} has been added to the inventory.")

    def remove_beer(self, name):
        for beer in self.beers:
            if beer.name.lower() == name.lower():
                self.beers.remove(beer)
                print(f"{name} has been removed from the inventory.")
                return
        print(f"{name} was not found in the inventory.")

    def update_quantity(self, name, new_quantity):
        for beer in self.beers:
            if beer.name.lower() == name.lower():
                beer.quantity = new_quantity
                print(f"{name}'s quantity updated to {new_quantity}.")
                return
        print(f"{name} was not found in the inventory.")

    def list_inventory(self):
        if not self.beers:
            print("The inventory is empty.")
        else:
            print("Current Beer Inventory:")
            for beer in self.beers:
                print(beer)

# Function to show menu
def show_menu():
    print("\nBeer Inventory Management")
    print("1. Add Beer")
    print("2. Remove Beer")
    print("3. Update Beer Quantity")
    print("4. List Inventory")
    print("5. Exit")

# Main
def run_inventory_system():
    inventory = Inventory()
    while True:
        show_menu()
        choice = input("Choose an option (1-5): ").strip()
        if choice == "1":
            name = input("Enter beer name: ").strip()
            style = input("Enter beer style: ").strip()
            try:
                quantity = int(input("Enter quantity: "))
                price = float(input("Enter price per unit: "))
            except ValueError:
                print("Invalid number entered.")
                continue
            beer = Beer(name, style, quantity, price)
            inventory.add_beer(beer)
        elif choice == "2":
            name = input("Enter the name of the beer to remove: ").strip()
            inventory.remove_beer(name)
        elif choice == "3":
            name = input("Enter the name of the beer to update: ").strip()
            try:
                new_quantity = int(input("Enter the new quantity: "))
            except ValueError:
                print("Invalid number entered.")
                continue
            inventory.update_quantity(name, new_quantity)
        elif choice == "4":
            inventory.list_inventory()
        elif choice == "5":
            print("Exiting the inventory system.")
            break
        else:
            print("Invalid choice. Please select an option between 1 and 5.")

run_inventory_system()
```

```python
import numpy as np
import time
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
```

```python
def insert_sort(a):
    for j in range(1,len(a)):
        val = a[j]
        # Inserta a[j] en la lista
        # ordenada a[0:j-1]
        i = j-1
        while (i>=0) and (a[i]>val):
            a[i+1] = a[i]
            i = i-1
        a[i+1] = val
```

```python
l_timings = []
for i, size in enumerate(range(5, 15)):
    t = list(reversed(range((size**2))))
    timings = %timeit -n 100 -r 10 -o -q insert_sort(t)
    l_timings.append([len(t), timings.best])
    a_timings = np.array(l_timings)
a_timings = np.array(l_timings)
_ = plt.plot(a_timings[ : , 0], a_timings[ : , 1]/a_timings[0, 1], '*')
```

```python
#Ajustar la función
def tofit(x, a, b):
    return a*x**2 + b

x = a_timings[: , 0]
y = np.array(a_timings)[: , 1]
y = y / y[0] #normalizar timings
pars, _ = curve_fit(tofit, x, y)
fit_insert_sort = pars[0]*(x**2) + pars[1]
print(fit_insert_sort)

plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, fit_insert_sort, label="y=a*x^2+b")

plt.show()
```

```python
def bubble_sort(a):
    n = len(a)
    for i in range(0,n-1):
        for j in range(n-1, i, -1):
            if(a[j-1]>a[j]):
                aux = a[j-1] 
                a[j-1] = a[j]
                a[j] = aux
```

```python
bubble_sort_l_timings = []
for i, size in enumerate(range(5, 15)):
    t = list(reversed(range((size**2))))
    timings = %timeit -n 100 -r 10 -o -q bubble_sort(t)
    bubble_sort_l_timings.append([len(t), timings.best])
    a_timings = np.array(bubble_sort_l_timings)
bubble_sort_a_timings = np.array(bubble_sort_l_timings)
_ = plt.plot(bubble_sort_a_timings[ : , 0], bubble_sort_a_timings[ : , 1]/bubble_sort_a_timings[0, 1], '*')
```

```python
#Ajustar la función
def bubble_sort_tofit(x, a, b):
    return a*x**2 + b

x = bubble_sort_a_timings[: , 0]
y = np.array(bubble_sort_a_timings)[: , 1]
y = y / y[0] #normalizar timings
pars, _ = curve_fit(bubble_sort_tofit, x, y)
fit_insert_sort = pars[0]*(x**2) + pars[1]
print(fit_insert_sort)

plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, fit_insert_sort, label="y=a*x^2+b")

plt.show()
```
